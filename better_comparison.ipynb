```python
# imamo q in s in samo na njiju nas zanima kako dobro se nauƒçita optimalne strategije

def determine_ratio(result):
    a, b = 0,0
    for elt in result:
        a += elt[0]
        b += elt[1]
    if a != 0:
        return b / a
    else:
        return 0

q = Qlearning(0.5, 0.5, 0.2)
s = SarsaAlgorithm(0.1, 0.5, 0.2)

num_trainings = [100, 500, 1000, 2500, 5000, 10000, 15000, 20000, 50000, 100000]


RS, RQ = [], []
for i in range(10):
    resultsr, resultss =dict(), dict()
    for t in num_trainings:
        # p = generate_piles(5, 10)
        p = [10, 10, 10, 10, 10]
        qq = train_model(t, q, p)
        ss = train_model(t, s, p)
        ar = random_opponent(500, qq, p, optimal_moves=True)
        br = random_opponent(500, ss, p, optimal_moves=True)
        # ass = smart_opponent(500, qq, p, optimal_moves=True)
        resultr = determine_ratio(ar)
        results = determine_ratio(br)
        resultsr[t] = resultr
        resultss[t] = results
        # print(resultsr)
        # print(resultss)
    RQ.append(resultsr)
    RS.append(resultss)
    print("evo: " + str(RQ))
    # RQ.append(resultss)
```
